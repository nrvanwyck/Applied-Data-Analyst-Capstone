## Demo: How to work with database


Download database file.

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df_30_to_35 = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
df_30_to_35_median = df_30_to_35.groupby(by='Age').median()
df_30_to_35_median = df_30_to_35_median[['WorkWeekHrs', 'CodeRevHrs']]
df_30_to_35_median.plot.bar(stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
df_30_to_35 = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
df_30_to_35_median = df_30_to_35.groupby(by='Age').median()
df_30_to_35_median = df_30_to_35_median[['ConvertedComp']]
df_30_to_35_median.plot.line()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
df_main_branch = df.groupby(by='MainBranch').median()
df_main_branch = df_main_branch[['Age']]
df_main_branch.plot.barh()
```

```python
QUERY = """
SELECT *
FROM DevType
"""

df = pd.read_sql_query(QUERY,conn)

respondents = list(set(df['Respondent'].tolist()))
front_end_respondents = df[df['DevType'] == 'Developer, front-end']['Respondent'].tolist()
mobile_respondents = df[df['DevType'] == 'Developer, mobile']['Respondent'].tolist()
full_stack_respondents = df[df['DevType'] == 'Developer, full-stack']['Respondent'].tolist()
back_end_respondents = df[df['DevType'] == 'Developer, back-end']['Respondent'].tolist()
print(len(front_end_respondents) / len(respondents))
print(len(mobile_respondents) / len(respondents))
print(len(full_stack_respondents) / len(respondents))
print(len(back_end_respondents) / len(respondents))
```
